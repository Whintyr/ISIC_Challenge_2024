```python
import matplotlib.pyplot as plt

# Plot histograms for all columns
def plot_histograms(df, dtype_dict):
    for column in df.columns:
        plt.figure(figsize=(10, 5))
        
        if dtype_dict[column] in ['int64', 'float64']:
            df[column].hist(bins=30, edgecolor='k')
            plt.title(f'Histogram of {column}')
            plt.xlabel(column)
            plt.ylabel('Frequency')
        elif dtype_dict[column] == 'object':
            df[column].value_counts().plot(kind='bar', edgecolor='k')
            plt.title(f'Bar plot of {column}')
            plt.xlabel(column)
            plt.ylabel('Count')
        
        plt.tight_layout()
        plt.show()

# Assuming you have a DataFrame named 'df'
# df = pd.read_csv('your_data.csv', dtype=dtype_dict)
plot_histograms(train_df, dtype_dict)
```
